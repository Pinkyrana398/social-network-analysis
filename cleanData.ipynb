## Python Script to Organize and Clean LinkedIn Data Files
### This script helps manage a folder containing .csv and .xlsx files. It performs the following tasks:
### 1.Finds all .csv and .xlsx files in subfolders
### 2.Moves .csv files to the main folder
### 3.Converts .xlsx files to .csv and moves them to the main folder
### 4.Deletes all other files and folders, keeping only the final .csv files

```python
import os
import pandas as pd
import shutil

# 👇 main folder
main_folder = 'LinkedIn Data Public'

# Step 1: Search for .csv and .xlsx files
csv_xlsx_files = []
for root, dirs, files in os.walk(main_folder):
    for file in files:
        ext = os.path.splitext(file)[1].lower()
        if ext in ['.csv', '.xlsx']:
            full_path = os.path.join(root, file)
            csv_xlsx_files.append(full_path)

# Step 2: Move/Convert files to main folder
for file_path in csv_xlsx_files:
    file_name = os.path.basename(file_path)
    ext = os.path.splitext(file_name)[1].lower()
    dest_path = os.path.join(main_folder, file_name)

    try:
        if ext == '.csv':
            if file_path != dest_path:
                shutil.move(file_path, dest_path)
                print(f"✅ Moved CSV: {file_name}")

        elif ext == '.xlsx':
            df = pd.read_excel(file_path)
            csv_name = os.path.splitext(file_name)[0] + '.csv'
            csv_path = os.path.join(main_folder, csv_name)
            df.to_csv(csv_path, index=False)
            print(f"✅ Converted XLSX to CSV: {csv_name}")
    except Exception as e:
        print(f"❌ Error with {file_name}: {e}")

# Step 3: Clean up - Delete everything except final CSVs in main folder
for root, dirs, files in os.walk(main_folder):
    for file in files:
        if not file.lower().endswith('.csv'):
            try:
                os.remove(os.path.join(root, file))
                print(f"🗑️ Deleted file: {file}")
            except Exception as e:
                print(f"❌ Error deleting {file}: {e}")
    for dir in dirs:
        try:
            shutil.rmtree(os.path.join(root, dir))
            print(f"🗑️ Deleted folder: {dir}")
        except Exception as e:
            print(f"❌ Error deleting folder {dir}: {e}")

print("\n🎉 Done! All final CSV files are now in:", main_folder)
```

## Script to Rename CSV Files in a Folder 
### This Python script is designed to rename all .csv files in a given folder (LinkedIn Data Public) by cleaning and formatting the filenames.
### 1.Searches for .csv files in the main folder
### 2.Cleans the filenames by:
### -Removing the part before a dash (-) if present
### -Replacing underscores (_) with spaces
### -Capitalizing each word
### 3.Renames each file to the cleaned format 

```python
import os

# Folder path
main_folder = 'LinkedIn Data Public'

# Loop through all files in the folder
for file_name in os.listdir(main_folder):
    file_path = os.path.join(main_folder, file_name)

    # Check if it's a CSV file
    if os.path.isfile(file_path) and file_name.lower().endswith('.csv'):
        # Remove the .csv extension temporarily
        name_part = file_name[:-4]

        # If it contains a dash, take only the part after the dash
        if '-' in name_part:
            name_part = name_part.split('-', 1)[1]

        # Replace underscores with spaces
        name_part = name_part.replace('_', ' ')

        # Capitalize each word
        new_name_part = ' '.join(word.capitalize() for word in name_part.split())

        # Final new filename
        new_filename = new_name_part + '.csv'
        new_file_path = os.path.join(main_folder, new_filename)

        # Rename the file
        try:
            os.rename(file_path, new_file_path)
            print(f"✅ Renamed: {file_name} ➡️ {new_filename}")
        except Exception as e:
            print(f"❌ Error renaming {file_name}: {e}")

print("\n🎉 All files have been renamed successfully!")
```

## Script to Clean and Standardize CSV Filenames
### This Python script processes .csv files in a folder (LinkedIn Data Public) and renames them using a clean and consistent format:
### 1.Scans for .csv files in the specified folder
### 2. Cleans filenames by: 
### -Removing prefixes (before dashes or special characters)
### -Replacing underscores and extra spaces
### -Extracting only alphabetic words (ignoring numbers/symbols)
### 3.Formats the final name as:
### - Firstname Lastname.csv (if 2+ words found)
### -Firstname.csv (if only 1 word found)
### -Unknown.csv (if no name can be extracted)
### 4.Renames the files with the new cleaned name


```python
import os
import re

# Folder path
main_folder = 'LinkedIn Data Public'

# Helper function to clean and format the filename
def extract_name(filename):
    # Remove .csv extension
    name = filename[:-4]

    # Remove any prefix before dash or special characters
    if '-' in name:
        name = name.split('-', 1)[1]
    
    # Replace underscores and extra spaces with single space
    name = name.replace('_', ' ')
    name = re.sub(r'\s+', ' ', name).strip()

    # Extract only alphabetic words (remove numbers, special chars)
    words = re.findall(r'[A-Za-z]+', name)

    # If we get at least 2 words, take first two as Firstname Lastname
    if len(words) >= 2:
        cleaned_name = f"{words[0].capitalize()} {words[1].capitalize()}"
    elif len(words) == 1:
        cleaned_name = words[0].capitalize()
    else:
        cleaned_name = "Unknown"

    return cleaned_name + ".csv"

# Loop through files and rename
for file in os.listdir(main_folder):
    file_path = os.path.join(main_folder, file)

    if os.path.isfile(file_path) and file.lower().endswith('.csv'):
        new_name = extract_name(file)
        new_path = os.path.join(main_folder, new_name)

        try:
            os.rename(file_path, new_path)
            print(f"✅ Renamed: {file} ➡️ {new_name}")
        except Exception as e:
            print(f"❌ Error renaming {file}: {e}")

print("\n🎯 Done! All files are now in 'Firstname Lastname.csv' format.")
```

## CSV Filename Cleaner & Duplicate Remover
### 1.Renames .csv files by:
### -Removing prefixes (e.g., before a -)
### -Replacing underscores (_) and extra spaces with single spaces
### -Extracting alphabetic words only
### -Capitalizing first two words as Firstname Lastname.csv
### 2.Checks for duplicates:
### -If two files would become the same after renaming, only one is kept
### -The second one is deleted



```python
import os
import re

main_folder = 'LinkedIn Data Public'

def extract_name(filename):
    name = filename[:-4]  # remove .csv
    if '-' in name:
        name = name.split('-', 1)[1]
    name = name.replace('_', ' ')
    name = re.sub(r'\s+', ' ', name).strip()
    words = re.findall(r'[A-Za-z]+', name)

    if len(words) >= 2:
        cleaned_name = f"{words[0].capitalize()} {words[1].capitalize()}"
    elif len(words) == 1:
        cleaned_name = words[0].capitalize()
    else:
        cleaned_name = "Unknown"

    return cleaned_name + ".csv"

# Track renamed files
for file in os.listdir(main_folder):
    file_path = os.path.join(main_folder, file)

    if os.path.isfile(file_path) and file.lower().endswith('.csv'):
        clean_name = extract_name(file)
        new_path = os.path.join(main_folder, clean_name)

        if os.path.abspath(file_path) == os.path.abspath(new_path):
            continue  # already correctly named

        if os.path.exists(new_path):
            # If file with target name already exists, delete this one
            os.remove(file_path)
            print(f"🗑️ Deleted duplicate: {file}")
        else:
            try:
                os.rename(file_path, new_path)
                print(f"✅ Renamed: {file} ➡️ {clean_name}")
            except Exception as e:
                print(f"❌ Error renaming {file}: {e}")

print("\n🎉 Done! All files renamed. Duplicates removed.")
```

## Clean & Format LinkedIn CSV Files
### This Python script processes .csv files in the LinkedIn Data Public folder.
### -It standardizes names into a single fullname column, removes invalid rows, and ensures clean data formatting.



```python
import os
import pandas as pd

# Path to your folder
folder = 'LinkedIn Data Public'

# Loop through all CSV files in the folder
for filename in os.listdir(folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder, filename)

        try:
            df = pd.read_csv(file_path)

            # Ensure firstname and lastname columns exist
            if 'firstname' in df.columns and 'lastname' in df.columns:
                # Fill NaNs with empty string and convert to string
                df['firstname'] = df['firstname'].fillna('').astype(str).str.strip()
                df['lastname'] = df['lastname'].fillna('').astype(str).str.strip()

                # Remove rows where both firstname and lastname are empty
                df = df[~((df['firstname'] == '') & (df['lastname'] == ''))]

                # Create fullname based on conditions
                df['fullname'] = df['firstname']
                df.loc[df['lastname'] != '', 'fullname'] = df['firstname'] + ' ' + df['lastname']

                # Remove unwanted characters, allow only a-z, A-Z, and space
                df['fullname'] = df['fullname'].str.replace(r'[^a-zA-Z ]+', '', regex=True)

                # Reorder columns: fullname first, then rest (excluding firstname and lastname)
                other_cols = [col for col in df.columns if col not in ['firstname', 'lastname', 'fullname']]
                df = df[['fullname'] + other_cols]

                # Save back to the same file
                df.to_csv(file_path, index=False)
                print(f"✅ Cleaned: {filename}")
            else:
                print(f"⚠️ Skipped (missing columns): {filename}")

        except Exception as e:
            print(f"❌ Error processing {filename}: {e}")
```

## Clean LinkedIn CSV Files with Fullname Generation
### This script processes .csv files in the LinkedIn Data Public folder. It adds a Fullname column, cleans the data, and ensures proper formatting.
### -Creates a "Fullname" column by combining "First Name" and "Last Name"
### -Removes rows where both first and last names are blank

```python
import os
import pandas as pd

# Folder containing the CSV files
folder_path = "LinkedIn Data Public"

# Loop through all CSV files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".csv"):
        file_path = os.path.join(folder_path, file_name)
        try:
            df = pd.read_csv(file_path, encoding='utf-8')

            # Check if "First Name" column exists
            if "First Name" in df.columns:
                # Replace NaN with empty string
                df["First Name"] = df["First Name"].fillna("")
                df["Last Name"] = df["Last Name"].fillna("") if "Last Name" in df.columns else ""

                # Strip whitespace and filter rows with both first and last names missing
                df["First Name"] = df["First Name"].astype(str).str.strip()
                df["Last Name"] = df["Last Name"].astype(str).str.strip()

                # Remove rows where both are empty
                df = df[~((df["First Name"] == "") & (df["Last Name"] == ""))]

                # Create "Fullname" column
                df["Fullname"] = (df["First Name"] + " " + df["Last Name"]).str.strip()

                # Capitalize first letters
                df["Fullname"] = df["Fullname"].str.title()

                # Save the updated file
                df.to_csv(file_path, index=False, encoding='utf-8')
                print(f"✅ Processed: {file_name}")
            else:
                print(f"⚠️ Skipped (missing 'First Name' column): {file_name}")
        except Exception as e:
            print(f"❌ Error processing {file_name}: {e}")
```

## Final LinkedIn CSV Cleaner
### 1.Handles various encodings (utf-8, latin1)
### 2.Normalizes column names
### 3.Extracts and formats Full Name and Company

```python
import os
import pandas as pd

folder = "LinkedIn Data Public"

def read_csv_with_encoding(file_path):
    try:
        return pd.read_csv(file_path, encoding='utf-8')
    except UnicodeDecodeError:
        try:
            return pd.read_csv(file_path, encoding='latin1')
        except:
            return None

for file_name in os.listdir(folder):
    if not file_name.lower().endswith(".csv"):
        continue

    file_path = os.path.join(folder, file_name)

    df = read_csv_with_encoding(file_path)
    if df is None:
        print(f"❌ Could not read: {file_name}")
        continue

    # Strip columns and standardize column names
    df.columns = [col.strip() for col in df.columns]

    # Skip files missing required columns
    if 'First Name' not in df.columns and 'Last Name' not in df.columns:
        print(f"⚠️ Skipped (missing 'First Name' and 'Last Name'): {file_name}")
        continue

    # Handle missing columns gracefully
    df['First Name'] = df.get('First Name', '').astype(str).str.strip()
    df['Last Name'] = df.get('Last Name', '').astype(str).str.strip()
    df['Company'] = df.get('Company', '').astype(str).str.strip()

    # Merge into Full Name
    df['Full Name'] = (df['First Name'] + ' ' + df['Last Name']).str.strip()

    # Title case and remove extra spaces
    df['Full Name'] = df['Full Name'].str.title()

    # Remove rows where Full Name is empty or just space
    df = df[df['Full Name'].str.strip() != '']

    # Keep only Full Name and Company
    df = df[['Full Name', 'Company']]

    # Drop rows where both columns are empty
    df.dropna(how='all', inplace=True)

    # Save the cleaned file
    df.to_csv(file_path, index=False, encoding='utf-8')
    print(f"✅ Processed: {file_name}")

print("\n🎉 Done! All files are processed and cleaned.")
```

## Advanced LinkedIn CSV Cleaner
### This script processes all .csv files in the LinkedIn Data Public folder. 

```python
import os
import pandas as pd

folder = "LinkedIn Data Public"

def read_csv_with_encoding_and_fallback(file_path):
    encodings = ['utf-8', 'latin1']
    delimiters = [',', ';', '\t']
    
    for enc in encodings:
        for delim in delimiters:
            try:
                return pd.read_csv(file_path, encoding=enc, delimiter=delim, on_bad_lines='skip')
            except Exception:
                continue
    return None

for file_name in os.listdir(folder):
    if not file_name.lower().endswith(".csv"):
        continue

    file_path = os.path.join(folder, file_name)
    df = read_csv_with_encoding_and_fallback(file_path)

    if df is None:
        print(f"❌ Could not read: {file_name}")
        continue

    df.columns = [col.strip() for col in df.columns]

    # Skip files with neither column
    if 'First Name' not in df.columns and 'Last Name' not in df.columns:
        print(f"⚠️ Skipped (missing 'First Name' and 'Last Name'): {file_name}")
        continue

    df['First Name'] = df.get('First Name', '').astype(str).str.strip()
    df['Last Name'] = df.get('Last Name', '').astype(str).str.strip()
    df['Company'] = df.get('Company', '').astype(str).str.strip()

    # Create Full Name
    df['Full Name'] = (df['First Name'] + ' ' + df['Last Name']).str.strip().str.title()

    # Drop empty Full Name
    df = df[df['Full Name'].str.strip() != '']

    # Keep only Full Name and Company
    df = df[['Full Name', 'Company']]

    df.dropna(how='all', inplace=True)

    try:
        df.to_csv(file_path, index=False, encoding='utf-8')
        print(f"✅ Processed: {file_name}")
    except Exception as e:
        print(f"❌ Error writing {file_name}: {e}")

print("\n🎉 All done! Check your folder.")
```

## LinkedIn Data Cleaning Script
### This script processes CSV files by cleaning and structuring data. It handles different encodings and delimiters, detects headers, and creates a **Full Name** column by combining **First Name** and **Last Name**. The cleaned files contain only the **Full Name** and **Company** columns. It saves the processed files back in the same folder.

```python
import os
import pandas as pd

folder = "LinkedIn Data Public"

def read_csv_with_fallback(file_path):
    encodings = ['utf-8', 'latin1', 'ISO-8859-1']
    delimiters = [',', ';', '\t']
    for enc in encodings:
        for delim in delimiters:
            try:
                return pd.read_csv(file_path, encoding=enc, delimiter=delim, skip_blank_lines=True, on_bad_lines='skip')
            except Exception:
                continue
    return None

def find_header_index(df):
    for i in range(min(len(df), 10)):
        row = df.iloc[i].astype(str).str.lower().tolist()
        if any("first name" in cell for cell in row):
            return i
    return -1

for file_name in os.listdir(folder):
    if not file_name.lower().endswith(".csv"):
        continue

    file_path = os.path.join(folder, file_name)
    df_raw = read_csv_with_fallback(file_path)

    if df_raw is None:
        print(f"❌ Failed to read: {file_name}")
        continue

    header_idx = find_header_index(df_raw)
    if header_idx == -1:
        print(f"⚠️ Header not found in: {file_name}")
        continue

    df = df_raw.iloc[header_idx:].reset_index(drop=True)
    df.columns = df.iloc[0].str.strip()
    df = df[1:].reset_index(drop=True)

    # Clean column names
    df.columns = [str(col).strip() for col in df.columns]

    # Check and extract relevant columns
    fname_col = next((col for col in df.columns if col.lower() == 'first name'), None)
    lname_col = next((col for col in df.columns if col.lower() == 'last name'), None)
    company_col = next((col for col in df.columns if 'company' in col.lower()), None)

    if fname_col is None and lname_col is None:
        print(f"⚠️ Skipped (no name columns): {file_name}")
        continue

    # Extract values and fill missing with empty string
    first_names = df[fname_col] if fname_col else ""
    last_names = df[lname_col] if lname_col else ""
    companies = df[company_col] if company_col else ""

    # Handle nulls and whitespace
    df['First Name'] = first_names.fillna('').astype(str).str.strip()
    df['Last Name'] = last_names.fillna('').astype(str).str.strip()
    df['Company'] = companies.fillna('').astype(str).str.strip()

    # Merge names and filter valid rows
    df['Full Name'] = (df['First Name'] + ' ' + df['Last Name']).str.strip().str.title()
    df = df[df['Full Name'].str.replace(" ", "") != ""]  # remove rows with no name

    # Final DataFrame with only required columns
    df_final = df[['Full Name', 'Company']]

    # Save back to same file
    try:
        df_final.to_csv(file_path, index=False, encoding='utf-8')
        print(f"✅ Cleaned: {file_name}")
    except Exception as e:
        print(f"❌ Error writing {file_name}: {e}")

print("\n🎯 Done! All files processed and cleaned.")
```
